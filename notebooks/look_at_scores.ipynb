```python
import matplotlib.pyplot as plt 
import numpy as np 

def plt_points(X,y,title=''):
    plt.rcParams["figure.figsize"] = (8,3)
    plt.stairs(*np.histogram(X.numpy()[y==0], bins=25), label='Benign')
    plt.stairs(*np.histogram(X.numpy()[y==1], bins=25), color='red', label='Malicious')
    
    plt.tight_layout()
    plt.legend()
    plt.title(title)
    plt.show()
```

```python
import torch 
sc = torch.load('scores/2.pt')
plt_points(*sc)
```
